```python
"""
Projekt ISS 2024/25: Podle hlasu poznáte je...

Autor: Václav Paťorek (xpator00)
Datum: 16. prosince 2024
"""
import numpy as np
from scipy import signal
import scipy
import warnings
warnings.filterwarnings('ignore')

def najdi_zakladni_frekvenci(signal, Fs, min_frekvence=30, max_frekvence=400):
    """
    Najde základní frekvenci v signálu pomocí FFT v rozsahu 30-400 Hz.
    Vrací frekvenci s nejvyšší amplitudou v daném rozsahu.
    """
    delka = len(signal)
    freq_spektrum = np.abs(np.fft.fft(signal))
    frekvence = np.fft.fftfreq(delka, 1/Fs)

    maska = (frekvence >= min_frekvence) & (frekvence <= max_frekvence)
    spektrum_v_rozsahu = freq_spektrum[maska]
    spicky = scipy.signal.find_peaks(spektrum_v_rozsahu, height=np.max(spektrum_v_rozsahu)*0.1)[0]

    if len(spicky) > 0:
        return frekvence[maska][spicky[0]]
    return None

def porovnej_spektra(signal1, signal2, Fs):
    """
    Porovná spektra dvou signálů a vrátí jejich podobnost.
    Signály jsou nejdřív normalizovány pro lepší porovnání.
    Vrací hodnotu 0-1 (1 = identická spektra)
    """
    delka = min(len(signal1), len(signal2))
    signal1 = signal1[:delka]
    signal2 = signal2[:delka]

    okno = np.hanning(delka)
    spektrum1 = np.abs(np.fft.fft(signal1 * okno))
    spektrum2 = np.abs(np.fft.fft(signal2 * okno))

    spektrum1 = spektrum1 / np.max(spektrum1)
    spektrum2 = spektrum2 / np.max(spektrum2)

    pul_delky = delka // 2
    podobnost = 1 - np.mean(np.abs(spektrum1[:pul_delky] - spektrum2[:pul_delky]))

    return podobnost
def normalizuj_harmonicke(signal, Fs, zakladni_frekvence=None):
    """
    Normalizuje harmonické složky signálu vzhledem k základní frekvenci.
    Vytáhne prvních 15 harmonických složek a znormalizuje jejich amplitudy.
    """
    if zakladni_frekvence is None:
        zakladni_frekvence = najdi_zakladni_frekvenci(signal, Fs)
        if zakladni_frekvence is None:
            return None

    delka = len(signal)
    spektrum = np.abs(np.fft.fft(signal))
    frekvence = np.fft.fftfreq(delka, 1/Fs)

    harmonicke_slozky = np.zeros(15)
    for i in range(len(harmonicke_slozky)):
        cilova_frekvence = zakladni_frekvence * (i+1)
        index = np.argmin(np.abs(frekvence - cilova_frekvence))
        harmonicke_slozky[i] = spektrum[index]

    return harmonicke_slozky / np.max(harmonicke_slozky)

def analyzuj_zvuk_motoru(signal, Fs):
    """
    Udělá kompletní analýzu zvukového signálu motoru:
    1. Najde základní frekvenci
    2. Spočítá harmonické složky
    3. Vypočítá spektrální těžiště
    """
    zakladni_frekvence = najdi_zakladni_frekvenci(signal, Fs)
    harmonicke = normalizuj_harmonicke(signal, Fs, zakladni_frekvence)

    frekvence, cas, spektrogram = scipy.signal.spectrogram(signal, Fs, nperseg=1024, noverlap=512)
    teziste = np.sum(frekvence[:, np.newaxis] * spektrogram, axis=0) / np.sum(spektrogram, axis=0)

    return {
        'zakladni_frekvence': zakladni_frekvence,
        'harmonicke': harmonicke,
        'spektralni_teziste': np.mean(teziste)
    }
def vypocitej_shodu_motoru(charakteristiky1, charakteristiky2, signal1, signal2, Fs):
    """
    Spočítá podobnost mezi dvěma zvuky motorů pomocí čtyř metrik:
    1. Podobnost harmonických složek (váha 30%)
    2. Podobnost spektrálního těžiště (váha 20%)
    3. Podobnost základní frekvence (váha 25%)
    4. Podobnost celkových spekter (váha 25%)
    """
    if charakteristiky1['harmonicke'] is None or charakteristiky2['harmonicke'] is None:
        return 0, 0, 0, 0

    # Podobnost harmonických
    harm_shoda = 1 - np.mean(np.abs(charakteristiky1['harmonicke'] - charakteristiky2['harmonicke']))

    # Podobnost těžiště
    teziste_shoda = 1 - np.abs(charakteristiky1['spektralni_teziste'] - charakteristiky2['spektralni_teziste']) / max(charakteristiky1['spektralni_teziste'], charakteristiky2['spektralni_teziste'])

    # Podobnost základní frekvence
    if charakteristiky1['zakladni_frekvence'] and charakteristiky2['zakladni_frekvence']:
        frekvencni_pomer = min(charakteristiky1['zakladni_frekvence'], charakteristiky2['zakladni_frekvence']) / max(charakteristiky1['zakladni_frekvence'], charakteristiky2['zakladni_frekvence'])
        frekvencni_chyba = np.min([np.abs(frekvencni_pomer - round(frekvencni_pomer)) for i in range(1, 4)])
        frekvencni_shoda = 1 - frekvencni_chyba
    else:
        frekvencni_shoda = 0

    # Podobnost celých spekter
    spektralni_shoda = porovnej_spektra(signal1, signal2, Fs)

    return harm_shoda, teziste_shoda, frekvencni_shoda, spektralni_shoda
def analyzuj_nahravky(test_signals, ref_signals, test_labels, ref_labels, Fs):
    """
    Hlavní funkce pro přiřazení testovacích nahrávek k referenčním.
    Projde všechny testovací nahrávky a pokusí se je přiřadit k referenčním.
    """
    pocet_testovacich = len(test_signals)
    pocet_referencnich = len(ref_signals)

    # Spočítám charakteristiky všech signálů
    referencni_charakteristiky = [analyzuj_zvuk_motoru(signal, Fs) for signal in ref_signals]
    testovaci_charakteristiky = [analyzuj_zvuk_motoru(signal, Fs) for signal in test_signals]

    # Vytvořím matici podobností
    matice_podobnosti = np.zeros((pocet_testovacich, pocet_referencnich))

    # Nastavím váhy pro jednotlivé typy podobností
    vahy = {
        'harmonicka': 0.30,  # 30% váha pro harmonické složky
        'teziste': 0.20,     # 20% váha pro spektrální těžiště
        'frekvencni': 0.25,  # 25% váha pro základní frekvenci
        'spektralni': 0.25   # 25% váha pro celkové spektrum
    }

    # Spočítám podobnosti mezi všemi dvojicemi
    for i in range(pocet_testovacich):
        for j in range(pocet_referencnich):
            harm_shoda, teziste_shoda, frekvencni_shoda, spektralni_shoda = vypocitej_shodu_motoru(testovaci_charakteristiky[i], referencni_charakteristiky[j], test_signals[i], ref_signals[j], Fs)

            # Spočítám váženou podobnost
            celkova_shoda = (harm_shoda * vahy['harmonicka'] + teziste_shoda * vahy['teziste'] + frekvencni_shoda * vahy['frekvencni'] + spektralni_shoda * vahy['spektralni'])
            matice_podobnosti[i, j] = celkova_shoda

    # Nastavím práh pro minimální podobnost
    MINIMALNI_PODOBNOST = 0.70
    pouzite_reference = set()

    # Pro každou testovací nahrávku najdu nejlepší shodu
    for i in range(pocet_testovacich):
        shody = [(j, matice_podobnosti[i,j]) for j in range(pocet_referencnich)]
        serazene_shody = sorted(shody, key=lambda x: x[1], reverse=True)

        # Projdu seřazené shody a najdu první nepřiřazenou referenci s dostatečnou podobností
        for j, shoda in serazene_shody:
            if ref_labels[j] not in pouzite_reference and shoda > MINIMALNI_PODOBNOST:
                pouzite_reference.add(ref_labels[j])
                print(f"Nahrávka {test_labels[i]} patří k nahrávce {ref_labels[j]} "
                      f"(podobnost {shoda*100:.1f}%)")
                break
        else:
            print(f"Nahrávka {test_labels[i]} nepatří k žádnému známému autu")

# Spuštění analýzy
analyzuj_nahravky(test_signals, ref_signals, test_labels, ref_labels, Fs)
```
